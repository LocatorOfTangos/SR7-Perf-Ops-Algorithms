```python
import json
import matplotlib.pyplot as plt

def getHeights(data_file):
    path = []
    with open(data_file, "r") as f:
        data = json.load(f)
        for chunk in data[1:]:
            path.append(chunk["height"])
    return path


shnoop = getHeights("wr_heights.json")
plt.plot(shnoop)
```
